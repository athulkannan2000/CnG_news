```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np

def get_articles(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    html_text = requests.get(url, headers=headers)
    soup = BeautifulSoup(html_text.content, 'html.parser')

    act_articles = soup.find_all('div', class_='col-md-6')
    result = []

    start_index = get_articles.previous_index if hasattr(get_articles, 'previous_index') else 0

    for idx, article in enumerate(act_articles, start=start_index + 1):
        act_title_tag = article.find('h1', class_='page-title')
        act_content_tag = article.find('div', class_='para-style1')

        act_title = act_title_tag.text if act_title_tag else None
        act_content = act_content_tag.text.strip() if act_content_tag else None
        act_htmlcontent = act_content_tag if act_content_tag else None

        article_data = {
            'Index': idx,
            'Title': act_title,
            'Link': url,
            'Content': act_content,
            'HTML Content': act_htmlcontent,
        }

        result.append(article_data)

    get_articles.previous_index = idx  # Store the last index for the next URL
    return result

# List of all the URLs
urls = [
    'http://www.autism-india.org/about-autism.php',
    'http://www.autism-india.org/identifying-autism.php',
    'http://www.autism-india.org/related-conditions.php',
    'http://www.autism-india.org/choosing-intervention.php',
    'http://www.autism-india.org/autism-india.php',
    'http://www.autism-india.org/history-autism-india.php',
    'http://www.autism-india.org/legislation-and-policies.php',
    'http://www.autism-india.org/self-advocacy-programme.php',
    'http://www.autism-india.org/inclusive-education.php',
    'http://www.autism-india.org/professional-awareness.php',
    'http://www.autism-india.org/media.php'
]

# Collecting data from each URL
all_articles_data = []

for url in urls:
    articles_data = get_articles(url)
    all_articles_data.extend(articles_data)

# Creating a DataFrame
df_articles = pd.DataFrame(all_articles_data)

# Printing the DataFrame
print(df_articles)
```

```python
df_articles.to_csv('ActionforAutism.csv', index = False)
```
